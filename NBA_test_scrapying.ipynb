```python
from selenium import webdriver
from bs4 import BeautifulSoup as Soup
import pandas as pd
from time import sleep
```

```python
driver = webdriver.Chrome("/Users/jessietsai/Desktop/data_science/chromedriver")
```

```python
driver.get('https://www.basketball-reference.com/')
```

```python
players = ['https://www.basketball-reference.com/players/j/jamesle01.html',
          'https://www.basketball-reference.com/players/d/duranke01.html',
          'https://www.basketball-reference.com/players/e/embiijo01.html',
          'https://www.basketball-reference.com/players/h/hardeja01.html',
          'https://www.basketball-reference.com/players/c/curryst01.html',
          'https://www.basketball-reference.com/players/d/doncilu01.html',
          'https://www.basketball-reference.com/players/e/embiijo01.html',
          'https://www.basketball-reference.com/players/i/irvinky01.html',
          'https://www.basketball-reference.com/players/p/paulch01.html',
          'https://www.basketball-reference.com/players/l/lillada01.html',
          'https://www.basketball-reference.com/players/j/jokicni01.html',]
```

```python
player_list = []
print('...loading...')
for i in range(11):
    driver.get(players[i])
    html = driver.page_source
    playersoup = Soup(html,'lxml')
    average = playersoup.find(id='per_game.2021')

    row = []
    player_name = driver.find_element_by_xpath('//*[@id="meta"]/div[2]/h1/span')
    row.append(player_name.text)
    for item in average.find_all('td'):
        row_item = item.text
        row.append(row_item)
    player_list.append(row)
print('---done---')
```

```python
df = pd.DataFrame(player_list)
df
```
