```python
from selenium import webdriver
from bs4 import BeautifulSoup as Soup
import pandas as pd
from time import sleep
```

```python
driver = webdriver.Chrome("/Users/jessietsai/Desktop/data_science/chromedriver")
```

```python
driver.get('https://www.basketball-reference.com/')
```

```python
#search for MVP
search = driver.find_element_by_name("search")
search.send_keys('MVP')

search_button = driver.find_element_by_xpath('//*[@id="header"]/div[3]/form/input[1]')
search_button.click()
```

```python
#2020 MVP page
mvp = driver.find_element_by_xpath('//*[@id="mvp_NBA"]/tbody/tr[2]/td[3]/a')
mvp.click()
```

```python
#scrape for top 10 MVP from 2020~2001
mvp_list = []
print('---start to scrape---')
for i in range(20):
    #解析頁面
    html = driver.page_source
    mvpsoup = Soup(html,'lxml')
    mvp_10 = mvpsoup.find(id='mvp') #find mvp table
    mvp_10 = mvp_10.find('tbody')
    mvp_10 = mvp_10.find_all('tr') #all rows
    
    #start scrape
    for r in range(10):
        row = [] #each row
        row.append(r+1)
        for item in mvp_10[r].find_all('td'): #find all items in a row
            row_item = item.text #tranform the item to text
            row.append(row_item)
        mvp_list.append(row)
    
    #get to last year page
    top = 'window.scrollTo(0,0)'
    driver.execute_script(top)
    pre_page = driver.find_element_by_xpath('//*[@id="content"]/div[1]/a[1]')
    pre_page.click()
    #last_year = driver.get('https://www.basketball-reference.com/awards/awards_'+k+'.html#mvp')
    sleep(2)
    print(len(mvp_list))
print('---end to scrape---')
```

```python
df = pd.DataFrame(mvp_list)
df
```
